```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=sns.load_dataset('titanic')
```

```python
data.info()
```

```python
cnf=data
cnf=cnf.dropna()
cnf.isna().sum()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='sex', y='age', hue='survived',data=cnf)
plt.title('Distribution of Age by Gender and Survival Status')
plt.xlabel('Gender')
plt.ylabel('Age')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()
```

```python
sns.distplot(cnf[cnf['survived']==0]['age'],hist=False,color='red')
sns.distplot(cnf[cnf['survived']==1]['age'],hist=False,color='green')
```
