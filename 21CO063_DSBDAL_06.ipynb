```python
cnf.boxplot()
```

```python
X=cnf.drop(['Species'],axis=1)
Y=cnf['Species']
```

```python
from sklearn.model_selection import train_test_split
xtre,xtes,ytre,ytes=train_test_split(X,Y,test_size=2,random_state=0)
```

```python
from sklearn.naive_bayes import GaussianNB
gnb=GaussianNB()
model=gnb.fit(xtre,ytre)
model
```

```python
ypred=gnb.predict(xtes)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix, precision_score, recall_score, f1_score,classification_report
```

```python
cm=confusion_matrix(ytes,ypred)
cm
```

```python
tp=cm[1,1]
fp=cm[0,1]
tn=cm[1,0]
fn=cm[0,0]
```

```python
ps=precision_score(ytes,ypred,average='micro')
ps
```

```python
rs=recall_score(ytes,ypred,average='micro')
rs
```

```python
acc=accuracy_score(ytes,ypred)
acc
```

```python
fs=f1_score(ytes,ypred,average='micro')
fs
```

```python
error_rate=1-acc
error_rate
```
