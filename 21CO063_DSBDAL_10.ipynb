```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Create histograms
data.hist(figsize=(10, 8))
plt.suptitle('Histograms of Features in Iris Dataset')
plt.show()


```

```python
# Create boxplots
data.boxplot(figsize=(10, 8))
plt.title('Boxplots of Features in Iris Dataset')
plt.show()
```

```python
%matplotlib inline
fig,axes=plt.subplots(2,2,figsize=(15,10))
axes[0,0].hist(data['SepalLengthCm'])
axes[0,0].set_title('SepalLengthCm')
axes[1,0].hist(data['SepalWidthCm'])
axes[1,0].set_title('SepalWidthCm')
axes[0,1].hist(data['PetalLengthCm'])
axes[0,1].set_title('PetalLengthCm')
axes[1,1].hist(data['PetalWidthCm'])
axes[1,1].set_title('PetalWidthCm')
```

```python
dataplot=[data['SepalLengthCm'],data['SepalWidthCm'],data['PetalLengthCm'],data['PetalWidthCm']]
sns.set_style("whitegrid") # Creating a figure instance
fig = plt.figure(1, figsize=(12,8))
# Creating an axes instance
ax = fig.add_subplot(111) # Creating the boxplot
bp = ax.boxplot(dataplot);
```
