```python
X=cnf.drop(['Purchased'],axis=1)
Y=cnf['Purchased']
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtre,xtes,ytre,ytes=train_test_split(X,Y,test_size=0.3,random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression

```

```python
logreg=LogisticRegression()
model=logreg.fit(xtre,ytre)
```

```python
model
```

```python
ypred=logreg.predict(xtes)
ypredt=logreg.predict(xtre)
```
